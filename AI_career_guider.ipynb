```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import time

# ==========================================
# 🧠 PART 1: THE SYNTHETIC BRAIN (Data Logic)
# ==========================================

# PM NOTE: To make the AI "smart", we can't use random numbers.
# We must simulate "archetypes" based on High School backgrounds.
# This creates correlations for the Neural Network to learn.

def generate_smart_data(n=8000):
    data = []
    streams = ['Science', 'Commerce', 'Arts']

    for _ in range(n):
        # 1. Pick a background
        stream = np.random.choice(streams)

        # 2. Assign Traits based on Background (Bias injection)
        # Base scores (0-10) with some noise
        if stream == 'Science':
            vis = np.random.randint(4, 11)   # Likely higher visual/spatial
            verb = np.random.randint(2, 9)
            strat = np.random.randint(5, 11) # Logic
            creat = np.random.randint(2, 9)
            algo = np.random.randint(6, 11)  # Engineering
            insight = np.random.randint(5, 11)
        elif stream == 'Commerce':
            vis = np.random.randint(2, 9)
            verb = np.random.randint(5, 11)  # Communication
            strat = np.random.randint(7, 11) # Strategy/Game Theory
            creat = np.random.randint(3, 9)
            algo = np.random.randint(2, 8)
            insight = np.random.randint(6, 11) # Statistics/Econ
        else: # Arts
            vis = np.random.randint(5, 11)   # Design
            verb = np.random.randint(7, 11)  # Literature
            strat = np.random.randint(2, 9)
            creat = np.random.randint(7, 11) # Imagination
            algo = np.random.randint(1, 7)
            insight = np.random.randint(2, 8)

        # 3. Determine "Correct" Label based on highest traits
        # We prioritize traits to assign labels for training
        scores = {'Visual': vis, 'Verbal': verb, 'Strat': strat, 'Creat': creat, 'Algo': algo, 'Insight': insight}
        sorted_traits = sorted(scores.items(), key=lambda x: x[1], reverse=True)
        top_trait = sorted_traits[0][0]
        second_trait = sorted_traits[1][0]

        subject = "General AI"

        # Rule-based labeling for training data
        if top_trait == 'Visual' or (top_trait == 'Algo' and second_trait == 'Visual'):
            subject = "Computer Vision 👁️"
        elif top_trait == 'Verbal':
            subject = "NLP (Language AI) 🗣️"
        elif top_trait == 'Strat' or (stream == 'Commerce' and top_trait == 'Algo'):
            subject = "Reinforcement Learning ♟️"
        elif top_trait == 'Creat':
            subject = "Generative AI 🎨"
        elif top_trait == 'Algo':
            subject = "Machine Learning Ops ⚙️"
        elif top_trait == 'Insight':
            subject = "Data Science 📊"

        data.append([stream, vis, verb, strat, creat, algo, insight, subject])

    return pd.DataFrame(data, columns=['Stream','Visual','Verbal','Strat','Creat','Algo','Insight','Subject'])

# --- TRAINING PIPELINE ---
print("⚙️ [Backend] Initializing AI Model...")
df = generate_smart_data()

# Features: Stream (Categorical) + 6 Traits (Numerical)
X = df[['Stream', 'Visual', 'Verbal', 'Strat', 'Creat', 'Algo', 'Insight']]
y = df['Subject']

# Preprocessing: OneHotEncode 'Stream', Scale others
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), ['Stream']),
        ('num', StandardScaler(), ['Visual', 'Verbal', 'Strat', 'Creat', 'Algo', 'Insight'])
    ])

# Prepare Target
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
class_names = label_encoder.classes_

# Fit Preprocessor
X_processed = preprocessor.fit_transform(X)

# Build Neural Network
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_processed.shape[1],)),
    BatchNormalization(),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(len(class_names), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print("🧠 [Backend] Training Neural Network on 8,000 synthetic profiles...")
model.fit(X_processed, y_encoded, epochs=15, batch_size=32, verbose=0)
print("✅ [System] Model Ready.")

# ==========================================
# 🎨 PART 2: THE PRODUCT FRONTEND (UI)
# ==========================================

# Style Definitions
style_html = """
<style>
    @import url('https://fonts.googleapis.com/css2?family=Inter:wght@300;400;600;700&display=swap');

    .main-container { font-family: 'Inter', sans-serif; padding: 20px; max-width: 700px; color: #1e293b; }
    .h-title { font-size: 32px; font-weight: 700; color: #4f46e5; letter-spacing: -0.025em; margin-bottom: 5px; }
    .h-desc { font-size: 16px; color: #64748b; margin-bottom: 30px; }

    .q-box {
        background: #ffffff;
        padding: 20px;
        border-radius: 12px;
        margin-bottom: 20px;
        border: 1px solid #e2e8f0;
        box-shadow: 0 4px 6px -1px rgba(0, 0, 0, 0.1), 0 2px 4px -1px rgba(0, 0, 0, 0.06);
    }
    .q-box .widget-html {
        font-weight: 600;
        color: #1e293b;
        margin-bottom: 15px;
        font-size: 18px;
    }
    .q-box .widget-radiobuttons label {
        font-size: 15px;
        margin-bottom: 8px;
        color: #334155;
    }
    .q-box .widget-radiobuttons input[type='radio']:checked + span {
        font-weight: 600;
        color: #4f46e5;
    }

    .result-card {
        background: linear-gradient(135deg, #e0f2f7, #d1e8ef);
        padding: 30px;
        border-radius: 15px;
        border: 1px solid #a7d9ed;
        box-shadow: 0 10px 15px -3px rgba(0, 0, 0, 0.1), 0 4px 6px -2px rgba(0, 0, 0, 0.05);
        color: #1e3a8a;
        line-height: 1.6;
    }
    .result-card h2 { color: #1c4587; font-size: 28px; margin-bottom: 10px; }
    .result-card h1 { color: #4f46e5; font-size: 48px; margin-top: 10px; margin-bottom: 15px; }
    .result-card p { font-size: 16px; margin-bottom: 8px; }
    .result-card hr { border: none; border-top: 1px dashed #a7d9ed; margin: 25px 0; }
    .result-card ul { list-style-type: '🚀 '; padding-left: 20px; }
    .result-card li { margin-bottom: 8px; }

    .widget-button button {
        background-color: #4f46e5;
        color: white;
        border: none;
        border-radius: 8px;
        padding: 10px 20px;
        font-size: 16px;
        cursor: pointer;
        transition: background-color 0.3s ease;
    }
    .widget-button button:hover { background-color: #4338ca; }

    .widget-dropdown select, .widget-text input {
        border: 1px solid #cbd5e1;
        border-radius: 8px;
        padding: 8px 12px;
        font-size: 15px;
        color: #334155;
    }
</style>
"""
display(HTML(style_html))

# --- COMPONENTS ---

# Progress Bar
progress = widgets.IntProgress(value=0, min=0, max=3, description='Progress:', bar_style='info', layout=widgets.Layout(width='98%'))

# Helper to make clean questions
def make_radio(question, options, key_values):
    # key_values matches options index to a Trait string
    r = widgets.RadioButtons(options=options, description=question, layout=widgets.Layout(width='95%'), style={'description_width': 'initial'})
    r.trait_map = key_values # Store the trait mapping inside the widget object
    return widgets.VBox([widgets.HTML(f"<div class='q-box'><div class='widget-html'>{question}</div>"), r, widgets.HTML("</div>")])

# --- PAGE 1: ONBOARDING (Demographics) ---
header_p1 = widgets.HTML("<div class='h-title'>👋 Sup, Future AI Pro!</div><div class='h-desc'>Let's get the deets on your background. No cap.</div>")

inp_stream = widgets.Dropdown(
    options=['Science', 'Commerce', 'Arts'],
    description='HS Vibe:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

inp_prof = widgets.Text(
    description='Current Main Quest:',
    placeholder='e.g. Student, Accountant, Designer...',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

page1 = widgets.VBox([header_p1, inp_stream, inp_prof], layout=widgets.Layout(margin='20px 0px'))


# --- PAGE 2: COGNITIVE STYLE (Questions 1-7) ---
header_p2 = widgets.HTML("<div class='h-title'>🧠 Level Up: Brain Mode</div><div class='h-desc'>How does your brain hit different when it's processing info?</div>")

q1 = make_radio("1. When you're tryna learn something new, you prefer:",
               ["Visuals, charts, and TikToks (jk, videos)", "Diving deep into the docs (reading is fundamental)", "Just yolo-ing it and learning by doing"],
               ['Visual', 'Verbal', 'Algo'])

q2 = make_radio("2. In the group project squad, you're usually:",
               ["The main character (big ideas only)", "The spreadsheet guru (all about the numbers)", "The one who actually builds the thing (the executor)"],
               ['Creat', 'Insight', 'Algo'])

q3 = make_radio("3. A math problem hits different. You:",
               ["Spotting patterns like it's a scavenger hunt", "Making mental movies of it (visualize it)", "Coding a script to brute-force the answer, duh"],
               ['Insight', 'Visual', 'Algo'])

q4 = make_radio("4. Your faves games are all about:",
               ["Big brain strategy (chess, real-time strategy)", "Quick reflexes & insane graphics (FPS, racing)", "Epic plots & deep convos (RPGs, story games)"],
               ['Strat', 'Visual', 'Verbal'])

q5 = make_radio("5. Which tool looks less sus and more fun to learn?",
               ["Photoshop / Illustrator (for the visuals)", "Excel / SQL (for the data vibes)", "Python / Terminal (for the coding energy)"],
               ['Creat', 'Insight', 'Algo'])

q6 = make_radio("6. If AI could be your personal assistant for one thing, what would it be?",
               ["Ghostwriting my essays/emails", "Making my photos/videos look fire", "Organizing my whole life (finances, schedule)"],
               ['Verbal', 'Visual', 'Strat'])

q7 = make_radio("7. Pick your aesthetic:",
               ["Logic (make it make sense)", "Aesthetics (make it look good)", "Rhetoric (make them listen)"],
               ['Strat', 'Visual', 'Verbal'])

page2 = widgets.VBox([header_p2, q1, q2, q3, q4, q5, q6, q7])
page2.layout.display = 'none'

# --- PAGE 3: TECHNICAL & PERSONALITY (Questions 8-15) ---
header_p3 = widgets.HTML("<div class='h-title'>⚙️ Phase 2: Tech Vibe Check</div><div class='h-desc'>Let's find your main character energy in the tech world.</div>")

q8 = make_radio("8. You're into work that's:",
               ["Super structured & has a right/wrong answer", "Wildly open-ended & lets you flex your creativity", "Like being a detective (uncovering hidden clues)"],
               ['Algo', 'Creat', 'Insight'])

q9 = make_radio("9. What part of AI gives you main character syndrome?",
               ["Robots doing their thing in the real world", "Chatbots understanding sarcasm (fr fr)", "AI creating art from just text prompts"],
               ['Strat', 'Verbal', 'Creat'])

q10 = make_radio("10. Dealing with 'Messy Data' sounds:",
                ["Like a satisfying glow-up (cleaning is zen)", "Annoying (just let me build the models already)", "Low-key boring (I want results, not chores)"],
                ['Insight', 'Algo', 'Strat'])

q11 = make_radio("11. 24 hours to a hackathon. You're building:",
                ["A stock market predictor that actually works", "A face filter app that breaks the internet", "A story generator that writes your next novel"],
                ['Insight', 'Visual', 'Verbal'])

q12 = make_radio("12. Which word is your aesthetic?",
                ["Precise (no crumbs left)", "Imaginative (dream big)", "Strategic (big brain moves only)"],
                ['Algo', 'Creat', 'Strat'])

q13 = make_radio("13. High School Flashback: You were low-key best at:",
                ["Geometry/Physics (visuals & how things work)", "Languages/History (storytelling & culture)", "Calculus/Statistics (numbers & patterns)"],
                ['Visual', 'Verbal', 'Insight'])

q14 = make_radio("14. Code's bugging out. You:",
                ["Staring at the logic until it clicks (it's a staring contest)", "Hitting up Google like it's your bestie for the error message", "Printing everything to debug (CSI: Code Edition)"],
                ['Algo', 'Verbal', 'Insight'])

q15 = make_radio("15. Final vibe check. You want a career that:",
                ["Optimizes systems (making things run smooth af)", "Creates new things (being a pioneer)", "Uncovers truths (spilling the tea on data)"],
                ['Strat', 'Creat', 'Insight'])

page3 = widgets.VBox([header_p3, q8, q9, q10, q11, q12, q13, q14, q15])
page3.layout.display = 'none'

# --- NAVIGATION & LOGIC ---
btn_next = widgets.Button(description="Next Flex ➡️", button_style='info', layout=widgets.Layout(width='30%'))
btn_submit = widgets.Button(description="🔮 Manifest My AI Path", button_style='success', layout=widgets.Layout(width='40%'))
btn_submit.layout.display = 'none'
out_display = widgets.Output()

current_page = 1

def on_next(b):
    global current_page
    if current_page == 1:
        # Move to Page 2
        page1.layout.display = 'none'
        page2.layout.display = 'block'
        progress.value = 1
        current_page = 2
    elif current_page == 2:
        # Move to Page 3
        page2.layout.display = 'none'
        page3.layout.display = 'block'
        btn_next.layout.display = 'none'
        btn_submit.layout.display = 'block'
        progress.value = 2
        current_page = 3

def get_application_advice(ai_field, profession):
    advice = f"As a {profession}, you can totally leverage {ai_field} to level up your game. For instance:"

    profession_lower = profession.lower()

    if "student" in profession_lower:
        if "vision" in ai_field.lower():
            advice += " Start a cool project like building an AI photo filter for your selfies or object detection for your campus. That's resume gold!"
        elif "nlp" in ai_field.lower():
            advice += " Work on a text summarizer for your notes or a chatbot for your study group. Think smart, not hard."
        elif "data science" in ai_field.lower():
            advice += " Analyze school data (like grades vs. study habits) or track trends in your favorite games. Get those insights!"
        elif "generative" in ai_field.lower():
            advice += " Use AI to generate unique art for school projects or even code creative writing prompts. No cap, that's lit."
        elif "reinforcement" in ai_field.lower():
            advice += " Experiment with AI to play simple games like Tic-Tac-Toe or optimize a study schedule. Big brain moves."
        elif "machine learning ops" in ai_field.lower():
            advice += " Focus on deploying small models you build, understanding how to make them run smoothly. It's like being a digital mechanic."
        else:
            advice += " Look for entry-level projects that align with both your studies and AI. Every little bit counts!"
    elif "accountant" in profession_lower or "finance" in profession_lower:
        if "data science" in ai_field.lower():
            advice += " Automate repetitive tasks like data entry, predict financial market trends, or detect fraudulent transactions. Money moves!"
        elif "nlp" in ai_field.lower():
            advice += " Use AI to summarize financial reports or analyze sentiment in earnings call transcripts. Get the TLDR faster."
        elif "reinforcement" in ai_field.lower():
            advice += " Optimize investment strategies or resource allocation using AI. Playing 4D chess with numbers."
        else:
            advice += " Explore AI tools that can streamline financial analysis and reporting. Efficiency is key!"
    elif "designer" in profession_lower or "artist" in profession_lower:
        if "generative" in ai_field.lower():
            advice += " Use AI to create unique visual assets, generate new design concepts, or even automate parts of your creative process. Unleash your inner AI Picasso!"
        elif "vision" in ai_field.lower():
            advice += " Develop AI tools for image recognition in art, enhance photos automatically, or create AI-driven filters. Picture perfect, literally."
        elif "nlp" in ai_field.lower():
            advice += " Create AI to generate compelling ad copy or assist with branding through language. Words that hit different."
        else:
            advice += " Leverage AI for mood board generation, trend prediction in design, or personalizing user experiences. Stay ahead of the curve."
    elif "engineer" in profession_lower or "developer" in profession_lower:
        if "machine learning ops" in ai_field.lower():
            advice += " Focus on deploying, monitoring, and maintaining AI models in production. Making AI work in the real world is clutch."
        elif "reinforcement" in ai_field.lower():
            advice += " Design AI agents to optimize complex systems, from robotics to network traffic. Smart systems are the future."
        elif "computer vision" in ai_field.lower():
            advice += " Build AI for quality control in manufacturing, autonomous vehicle navigation, or even smart security systems. See it, build it."
        else:
            advice += " Integrate AI functionalities into existing software, optimize code with AI-driven insights, or build smart automation tools. Level up your code game."
    elif "sales" in profession_lower or "marketing" in profession_lower:
        if "nlp" in ai_field.lower():
            advice += " Use AI to analyze customer feedback, personalize marketing messages, or automate lead qualification. Speak to your audience on another level."
        elif "data science" in ai_field.lower():
            advice += " Predict customer behavior, optimize ad spend, or identify new market opportunities. Data is the new gold, fr."
        elif "generative" in ai_field.lower():
            advice += " Create AI-generated content for campaigns, personalized product descriptions, or innovative ad creatives. Stand out from the crowd."
        else:
            advice += " Implement AI for CRM integration, sales forecasting, or customer service automation. Work smarter, not harder."
    else:
        advice += f" Explore how {ai_field} can bring automation, intelligent decision-making, or creative solutions to your daily tasks and challenges. The possibilities are endless, bestie."

    return advice

def calculate_result(b):
    page3.layout.display = 'none'
    btn_submit.layout.display = 'none'
    progress.value = 3
    progress.bar_style = 'success'

    with out_display:
        clear_output()
        print("⚡ Getting your background data, fam...")
        time.sleep(0.5)
        print("⚡ Flexing on 15 data points...")
        time.sleep(0.5)
        print("⚡ Running the AI brain for your personalized prediction...")
        time.sleep(0.8)
        clear_output()

        # 1. Gather Scores
        user_stream = inp_stream.value
        user_prof = inp_prof.value

        # Initialize trait scores (Mid-point)
        scores = {'Visual':5, 'Verbal':5, 'Strat':5, 'Creat':5, 'Algo':5, 'Insight':5}

        all_questions = [q1, q2, q3, q4, q5, q6, q7, q8, q9, q10, q11, q12, q13, q14, q15]

        for q_vbox in all_questions: # Changed from 'q' to 'q_vbox' to reflect it's a VBox
            q = q_vbox.children[1] # The RadioButtons widget is the second child
            selected_index = q.index # Get index of selection (0, 1, or 2)
            trait_name = q.trait_map[selected_index] # Map to trait
            scores[trait_name] += 1.5 # Boost score

        # 2. Prepare Data Frame for Model
        input_df = pd.DataFrame([[
            user_stream,
            scores['Visual'], scores['Verbal'], scores['Strat'],
            scores['Creat'], scores['Algo'], scores['Insight']
        ]], columns=['Stream','Visual','Verbal','Strat','Creat','Algo','Insight'])

        # 3. Transform & Predict
        input_processed = preprocessor.transform(input_df)
        probs = model.predict(input_processed, verbose=0)

        # Get Top 2 Predictions
        top_indices = np.argsort(probs[0])[-2:][::-1]
        top_pred = class_names[top_indices[0]]
        runner_up = class_names[top_indices[1]]
        confidence = probs[0][top_indices[0]] * 100

        # Generate application advice
        application_advice = get_application_advice(top_pred, user_prof)

        # 4. Display "Product" Card
        display(HTML(f"""
        <div class='result-card'>
            <center>
                <h2>✨ YOUR AI CAREER VIBE CHECK ✨</h2>
                <h1 style='color:#4f46e5; font-size:48px;'>{top_pred}</h1>
                <p><b>Match Confidence:</b> {confidence:.1f}% (No cap!)</p>
                <hr>
            </center>

            <p><b>👤 Your Origin Story:</b> You were a <b>{user_stream}</b> main, now you're an expert <b>{user_prof}</b>.</p>
            <p><b>🥈 Another Strong Vibe:</b> {runner_up} (also a solid choice, ngl)</p>

            <br>
            <h3>🔍 Why this path slaps for you:</h3>
            <p>Based on your <b>{user_stream}</b> background and your quiz responses, your brain is totally tuned for
            <b>{max(scores, key=scores.get)}</b> and tech challenges. This is your moment.</p>

            <div style='background-color:#eff6ff; padding:15px; border-left:4px solid #6366f1; border-radius:8px;'>
                <b>🚀 Next Level Moves:</b>
                <ul>
                    <li><b>If {top_pred.split()[0]} is giving new energy:</b> Start with Python and {top_pred.split()[0]} basics on Coursera. It's a whole mood.</li>
                    <li><b>Project Idea:</b> {application_advice}</li>
                    <li><b>Tools to master:</b> {get_tools(top_pred)} (they're the meta)</li>
                </ul>
            </div>
        </div>
        """))

def get_tools(pred):
    if "Vision" in pred: return "OpenCV, PyTorch, YOLO (the holy trinity)"
    if "NLP" in pred: return "HuggingFace, Transformers, NLTK (for all the word nerds)"
    if "Data" in pred: return "Pandas, SQL, Scikit-Learn (data analysis on fleek)"
    if "Generative" in pred: return "Stable Diffusion, Midjourney, LangChain (unleash your inner creator)"
    if "Reinforcement" in pred: return "OpenAI Gym, Unity ML, Ray (for the strategic masterminds)"
    return "TensorFlow, Keras, Docker (the OG essentials)"

btn_next.on_click(on_next)
btn_submit.on_click(calculate_result)

# Run App
display(progress)
display(page1)
display(page2)
display(page3)
display(widgets.HBox([btn_next, btn_submit], layout=widgets.Layout(justify_content='center')))
display(out_display)
```
